# Perceptron Algorithm

### Importing modules

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

### Rough

```python
x = np.linspace(0, 10, 400)
x_points = np.linspace(0, 10, 50) 

# Define the corresponding y values for the line y = 3x + 2
y = 3 * x + 2
y_points = 3 * x_points + 2
class_1 = y_points + (np.random.rand(50)*10)
class_2 = y_points - (np.random.rand(50)*10)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 3x + 2', color='blue')
plt.scatter(x_points, class_1, label='Class 1', color='red')
plt.scatter(x_points, class_2, label='Class 2', color='green')


# Add axes, grid, and labels
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel('Tennis')
plt.ylabel('Painting')
plt.title('Original Classification')
plt.legend()
plt.show()
```

### Util

```python
def generate_data(num_points, orig_weights = [-3,1], orig_bias = -2, show_plot = True):
    x1_class1 = np.linspace(0, 10, num_points//2)
    x2_base_class1 = (-orig_weights[0]/orig_weights[1]) * x1_class1 + (-orig_bias/orig_weights[1])
    x2_class1 =  x2_base_class1 + (np.random.rand(num_points//2)*10)

    x1_class0 = np.linspace(0, 10, num_points - (num_points//2))
    x2_base_class0 = (-orig_weights[0]/orig_weights[1]) * x1_class0 + (-orig_bias/orig_weights[1])
    x2_class0 =  x2_base_class0 - (np.random.rand(num_points - (num_points//2))*10)

    x1 = np.concatenate((x1_class1, x1_class0))
    x2 = np.concatenate((x2_class1, x2_class0))
    
    data = np.vstack((x1, x2))
    labels = np.concatenate((np.ones(num_points//2), np.zeros(num_points - (num_points//2))))

    if show_plot:        
        plot(data, labels, weights = orig_weights, bias = orig_bias)
    
    
    return data, labels

def accuracy(true_labels, predicted_labels):
    return sum(true_labels == predicted_labels)/len(true_labels)
    

def plot(dataset, labels, weights = [-3,1], bias = -2, plot_original = True):

    plt.figure(figsize=(8, 6))
    x = np.linspace(0, 10, 400)
    y = -weights[0]/weights[1] * x - (bias/weights[1])  
    plt.plot(x, y, label='Classifier Line', color='blue')

    if plot_original:
        x = np.linspace(0, 10, 400)
        y = 3 * x +2
        plt.plot(x, y, label='Original Line', color='red')

    plt.scatter(dataset[0, labels == 0], dataset[1, labels == 0], label='Class 0', color='red')
    plt.scatter(dataset[0, labels == 1], dataset[1, labels == 1], label='Class 1', color='green')
    
    # Add axes, grid, and labels
    plt.axhline(0, color='black', linewidth=0.8)
    plt.axvline(0, color='black', linewidth=0.8)
    plt.xlim(0, 10)
    plt.ylim(-10, 40)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.xlabel('Tennis')
    plt.ylabel('Painting')
    plt.title('Classification')
    plt.legend()
    plt.show()
    
    return None
```

### Perceptron Class

```python
class Perceptron:
    def __init__(self, dimensions, alpha):
        self.dimensions = dimensions
        self.weights = np.random.rand(dimensions,1)
        self.bias = np.random.rand(1)
        self.alpha = alpha

    def fit(self, train_data, train_labels, epochs):
        
        for epoch in range(epochs):
            predicted_labels = self.predict(train_data)
            self.update_weights(train_labels, predicted_labels, train_data)
            # print("Weights, Bias: ", self.weights, self.bias)
            # print("Accuracy: ", accuracy(train_labels, predicted_labels))
            # plot(train_set, train_labels, self.weights, self.bias)

    def predict(self, data):
        score = self.weights.T @  data + self.bias
        score[score <= 0] = 0
        score[score > 0] = 1
        predicted_labels = score[0]

        return predicted_labels
        
        
    def update_weights(self, true_labels, predicted_labels, train_data):
        i=0
        for true, predict in zip(true_labels, predicted_labels):
            if true != predict:
                if predict ==0:
                    self.weights += (self.alpha*train_data[:, i].reshape(-1,1))
                    self.bias += self.alpha
                else:
                    self.weights -= self.alpha
                    self.bias -= self.alpha
            i+=1
        return None
```

### Main

```python
num_points = 100
dataset, labels = generate_data(num_points, show_plot = False)

train_set, test_set = dataset[:,:int(0.8*num_points)], dataset[:,int(0.8*num_points):]
train_labels, test_labels = labels[:int(0.8*num_points)], labels[int(0.8*num_points):]


#Initialisation
model= Perceptron(dimensions=2, alpha =0.3)
predictions = model.predict(test_set)
accuracy_score = accuracy(test_labels, predictions)
plot(train_set, train_labels, model.weights, model.bias)
print("Accuracy Score: ", accuracy_score)

#Training
model.fit(train_set, train_labels,epochs = 1000)
plot(train_set, train_labels, model.weights, model.bias)

predictions = model.predict(test_set)

accuracy_score = accuracy(test_labels, predictions)
print("Accuracy Score: ", accuracy_score)

```
